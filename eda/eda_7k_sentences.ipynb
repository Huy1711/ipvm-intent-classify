```python
data = []
for index, row in df.iterrows():
    text = row['question']
    label = [key for key in row.keys() if row[key] == 1][0]
    data.append({"text": text, "label": label})
```

```python
import json
with open("7k_sentences_train.jsonl", "w") as f:
    for sent in data:
        f.write(json.dumps(sent) + "\n")
```
